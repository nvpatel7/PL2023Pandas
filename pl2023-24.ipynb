```python
import pandas as pd
```

Reading in the csv file and printing the first few rows to understand the data.

```python
df = pd.read_csv('premier-player-23-24.csv')
df.head()
```

Getting the unique values of the 'class' column to understand the target variable.

```python
df.shape
df.info()
```

converting age to int

```python
df['Age'] = df['Age'].astype(int)
df.head()
```

convertin minutes to int

```python
df['Min']= df['Min'].astype(int)
df.head()
```

```python
df['Gls'] = df['Gls'].astype(int)
df['Ast'] = df['Ast'].astype(int)

df.head()
```

```python
df.drop(columns=['npxG+xAG', 'Ast_90', 'PrgP', 'PrgR', 'PrgC', 'Gls_90', 'npxG', 'npxG+xAG_90', 'npxG_90',
                 'xG+xAG_90', 'xAG_90', 'xG_90', 'G+A-PK_90', 'G-PK_90', 'G+A_90', 'G-PK', 'PKatt', 'PK'],
        inplace=True)

df.head()
```

```python
columns = ['G+A', 'CrdY', 'CrdR']

for col in columns:
    df[col] = df[col].astype(int)

df.head()
```

```python
nation = df['Nation'].str.split(' ').str[1]

df['Nation'] = nation
df.head()
```

```python
df['xG+xAG'] = df['xG'] + df['xAG']

df.head()
```

```python
# Get the list of all column names
cols = list(df.columns)

# Move the 'xG+xAG' column to be second to last
cols.insert(-1, cols.pop(cols.index('xG+xAG')))

# Reorder the DataFrame with the updated column order
df = df[cols]

df.head()
```

```python
df.isna().sum()
```

```python
df['Pos2'] = df['Pos'].str.split(',').str[1]

df['Pos2'] = df['Pos2'].fillna('None')


```

```python
df['Pos'] = df['Pos'].str.split(',').str[0]

df['Pos'] = df['Pos'].str.strip().fillna('None')

df
```

```python
cols = list(df.columns)

cols.insert(3, cols.pop(cols.index('Pos2')))

df = df[cols]

df.head()
```

```python
cols = list(df.columns)

cols.insert(-1, cols.pop(cols.index('CrdY')))
cols.insert(-1, cols.pop(cols.index('CrdR')))
cols.insert(2, cols.pop(cols.index('Team')))


df = df[cols]

df
```

```python
df.rename(columns={'Gls':'Goals', 'Ast':'Assists', 'xG':'Expected Goals', 
                   'xAG':'Expected Assists', 'CrdY':'Yellow Cards', 
                   'CrdR':'Red Cards', 'Player':'Name', 'Min': 'Minutes'}, inplace = True)
df
```

```python
df.reset_index(drop=True, inplace=True)

df.head()
```

# Challenge 1: Filter Top Scorers
**Task:** Filter the rows of players who have scored more than 10 goals so far in the season. Display their names and the number of goals.


```python
df.loc[df['Goals'] > 10, ['Name', 'Goals']].sort_values(by = 'Goals',ascending=False)
```

# Challenge 2: Find Players with High Assists and Low Goals
**Task:** Filter players who have more than 5 assists but fewer than 3 goals. Display their names, assists, and goals.

```python
HALG = df.loc[(df['Assists'] > 5) & (df['Goals'] < 3), ['Name', 'Assists', 'Goals']]

HALG.sort_values(by=['Assists', 'Goals'], ascending=False)
```

# Challenge 3: Filter Players by Position and High xG
**Task:** Filter midfielders who have an xG (expected goals) greater than 5. Display their names, position, and xG.


```python
highxG = df.loc[(df['Expected Goals'] > 5) & 
                ((df['Pos'] == 'MF') | (df['Pos2'] == 'MF')), 
                ['Name', 'Pos', 'Pos2', 'Expected Goals']]

highxG.sort_values(by='Expected Goals', ascending=False)
```

# Challenge 4: Players with High xAG but No Goals
**Task:** Filter players who have an xAG (expected assists) greater than 5 but haven't scored any goals. Display their names, xAG, and goals.


```python
high_xAG_no_goals = df.loc[(df['Expected Assists'] > 5) & (df['Goals'] == 0), ['Name', 'Expected Assists', 'Goals']]

high_xAG_no_goals.sort_values(by='Expected Assists', ascending=False)
```

# Challenge 5: Filter Players by Nationality
**Task:** Filter all players from a specific nationality (e.g., "English") who have more than 2 assists. Display their names, nationality, and assists.

```python
english_brazilian_players = df[((df['Nation'] == 'ENG') | (df['Nation'] == 'BRA')) & 
                     (df['Assists'] > 2)][['Name', 'Nation', 'Assists']]

english_brazilian_players.sort_values(by='Assists', ascending=False)
```

# Challenge 6: Filter by Minutes Played
**Task:** Filter players who have played more than 1000 minutes but have fewer than 5 total goal involvements (goals + assists). Display their names, minutes played, goals, and assists.


```python
low_involvement = df[(df['Minutes'] > 1000) & (df['G+A'] < 5)][['Name', 'Minutes', 'Goals', 'Assists']]

low_involvement.sort_values(by=['Goals', 'Assists'], ascending=False)
```

# Challenge 7: Filter Players with a High Contribution per 90 Minutes
**Task:** Filter players with goal involvements per 90 minutes greater than 0.5. First, calculate the goal involvement per 90 minutes, and then filter for players exceeding this value. Display their names and contribution per 90 minutes.


```python
df['goal_involvement_per_90'] = (df['G+A']) / (df['Minutes'] / 90)

high_involvement_players = df[df['goal_involvement_per_90'] > 0.5][['Name', 'goal_involvement_per_90']]

high_involvement_players.sort_values(by='goal_involvement_per_90', ascending=False)
```

# Challenge 8: Filter by Age Group
**Task:** Filter all players who are under the age of 23. Display their names, age, and goals.


```python
u23 = df[df['Age'] < 23][['Name', 'Age', 'Goals']]

u23.sort_values(by=['Goals', 'Age'], ascending=False)
```

# Challenge 9: Filter Players with Low xG but High Goals
**Task:** Find players who have scored more than 10 goals but have an xG (expected goals) of less than 8. Display their names, goals, and xG.

# Challenge 10: Filter Players with High xA and High Assists
